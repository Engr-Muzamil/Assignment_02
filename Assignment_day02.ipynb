# Library system with book and member classes

```python
from dataclasses import dataclass, field

@dataclass
class Book:
    book_id: str
    title: str
    author: str
    available: bool = True
    borrowed_by: str | None = None  # member_id

    def borrow(self, member_id: str) -> bool:
        if not self.available:
            return False
        self.available = False
        self.borrowed_by = member_id
        return True

    def return_book(self) -> bool:
        if self.available:
            return False
        self.available = True
        self.borrowed_by = None
        return True


@dataclass
class Member:
    member_id: str
    name: str
    borrowed_books: list[str] = field(default_factory=list)

    def borrow_book(self, book: Book) -> bool:
        if book.borrow(self.member_id):
            self.borrowed_books.append(book.book_id)
            return True
        return False

    def return_book(self, book: Book) -> bool:
        if book.book_id not in self.borrowed_books:
            return False
        if book.return_book():
            self.borrowed_books.remove(book.book_id)
            return True
        return False


class Library:
    def __init__(self):
        self.books: list[Book] = []
        self.members: list[Member] = []

    def add_book(self, book: Book) -> None:
        self.books.append(book)

    def add_member(self, member: Member) -> None:
        self.members.append(member)

    def find_book(self, book_id: str) -> Book | None:
        return next((b for b in self.books if b.book_id == book_id), None)

    def find_member(self, member_id: str) -> Member | None:
        return next((m for m in self.members if m.member_id == member_id), None)

    def borrow_book(self, member_id: str, book_id: str) -> bool:
        m, b = self.find_member(member_id), self.find_book(book_id)
        return bool(m and b and m.borrow_book(b))

    def return_book(self, member_id: str, book_id: str) -> bool:
        m, b = self.find_member(member_id), self.find_book(book_id)
        return bool(m and b and m.return_book(b))


if __name__ == "__main__":
    lib = Library()
    lib.add_book(Book("B1", "Clean Code", "Robert C. Martin"))
    lib.add_book(Book("B2", "Python Crash Course", "Eric Matthes"))
    lib.add_member(Member("M1", "Ali"))

    print(lib.borrow_book("M1", "B1"))   # True
    print(lib.borrow_book("M1", "B1"))   # False (already borrowed)
    print(lib.return_book("M1", "B1"))   # True
```

# Employee Payroll system

```python
from dataclasses import dataclass

@dataclass
class Employee:
    name: str
    base_salary: float

    def calculate_salary(self) -> float:
        return self.base_salary


@dataclass
class Manager(Employee):
    bonus: float = 0.0  # fixed bonus

    def calculate_salary(self) -> float:
        return super().calculate_salary() + self.bonus


@dataclass
class Engineer(Employee):
    overtime_hours: float = 0.0
    overtime_rate: float = 0.0
    performance_bonus: float = 0.0

    def calculate_salary(self) -> float:
        overtime_pay = self.overtime_hours * self.overtime_rate
        return super().calculate_salary() + overtime_pay + self.performance_bonus


class Payroll:
    def __init__(self):
        self.employees: list[Employee] = []

    def add_employee(self, emp: Employee) -> None:
        self.employees.append(emp)

    def total_payroll(self) -> float:
        return sum(e.calculate_salary() for e in self.employees)

    def payslips(self) -> list[tuple[str, float]]:
        return [(e.name, e.calculate_salary()) for e in self.employees]


if __name__ == "__main__":
    payroll = Payroll()
    payroll.add_employee(Employee("Ayesha", 50000))
    payroll.add_employee(Manager("Usman", 80000, bonus=15000))
    payroll.add_employee(Engineer("Sara", 70000, overtime_hours=10, overtime_rate=500, performance_bonus=3000))

    print(payroll.payslips())
    print("Total:", payroll.total_payroll())
```

# Contact Book App

```python
import json
from pathlib import Path

class ContactBook:
    def __init__(self, filename: str = "contacts.json"):
        self.filepath = Path(filename)
        self.contacts: dict[str, dict] = {}
        self.load()

    def load(self) -> None:
        if self.filepath.exists():
            self.contacts = json.loads(self.filepath.read_text(encoding="utf-8") or "{}")
        else:
            self.contacts = {}

    def save(self) -> None:
        self.filepath.write_text(json.dumps(self.contacts, indent=2), encoding="utf-8")

    def add(self, name: str, phone: str = "", email: str = "") -> None:
        self.contacts[name] = {"phone": phone, "email": email}
        self.save()

    def search(self, name: str) -> dict | None:
        return self.contacts.get(name)

    def delete(self, name: str) -> bool:
        if name in self.contacts:
            del self.contacts[name]
            self.save()
            return True
        return False

    def list_all(self) -> list[tuple[str, dict]]:
        return sorted(self.contacts.items(), key=lambda x: x[0].lower())


def menu():
    cb = ContactBook()

    while True:
        print("\n1) Add  2) Search  3) Delete  4) List All  5) Exit")
        choice = input("Choose: ").strip()

        if choice == "1":
            name = input("Name: ").strip()
            phone = input("Phone: ").strip()
            email = input("Email: ").strip()
            if name:
                cb.add(name, phone, email)
                print("Saved.")
            else:
                print("Name required.")

        elif choice == "2":
            name = input("Search name: ").strip()
            found = cb.search(name)
            print(found if found else "Not found.")

        elif choice == "3":
            name = input("Delete name: ").strip()
            print("Deleted." if cb.delete(name) else "Not found.")

        elif choice == "4":
            for name, info in cb.list_all():
                print(f"{name} -> phone: {info.get('phone','')}, email: {info.get('email','')}")

        elif choice == "5":
            break

        else:
            print("Invalid choice.")


if __name__ == "__main__":
    menu()
```
